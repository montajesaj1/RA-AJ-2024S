```python
import pandas as pd
```

```python
df_1 = pd.read_csv("iqvia_drugs_matched_to_pubchem.csv")
df_2 = pd.read_csv("PUBCHEM_SYNONYMS_FLAGGED.csv")
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df_2['cas_both'] = df_2['CAS'] + ', ' + df_2['Deprecated CAS']
df_2.drop(['CAS', 'Deprecated CAS'], axis=1, inplace=True)

# Splitting "cas_both" into separate rows
df_2 = df_2.drop('cas_both', axis=1).join(
    df_2['cas_both'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('cas_both')
).reset_index(drop=True)
```

```python
df_2 = df_2[df_2['cas_both'] != "No CAS info"].reset_index(drop=True)
```

```python
df_2.head(50)
```

```python
df_1.head(50)
```

```python
df_1_exploded = df_1.assign(cas=df_1['cas'].str.split(', ')).explode('cas')

# Map CAS to Parsed Molecule from df_2
cas_to_parsed_molecule = df_2.set_index('cas_both')['Parsed Molecule'].to_dict()
df_1_exploded['Parsed Molecule'] = df_1_exploded['cas'].map(cas_to_parsed_molecule)

# Group by original index and CAS to check if all Parsed Molecule values are the same for each CAS
def check_different_compounds(group):
    if len(set(group['Parsed Molecule'])) > 1:
        return 1
    else:
        return 0

df_1['different_compounds'] = df_1_exploded.groupby(level=0).apply(check_different_compounds).values

df_1.head(50)
```

```python
matched = df_1.to_csv("iqvia_drugs_matched_compounds_flagged.csv", index=False)
```
