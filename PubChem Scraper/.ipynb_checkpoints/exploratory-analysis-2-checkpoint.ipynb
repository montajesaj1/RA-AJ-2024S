```python
import pandas as pd
```

```python
correct_data = pd.read_csv("pubchem_with_eight.csv", index_col=0)
```

```python
old_data = pd.read_csv("PUBCHEM_SYNONYMS_FLAGGED.csv")
old_data
```

```python
merged_df = pd.merge(old_data, correct_data, how='inner')
merged_df
```

```python
merged_df.to_csv("unchanged_data.csv", index=False)
```

```python
merged = pd.merge(old_data, merged_df, on=['Original Molecule', 'Parsed Molecule'], how='left', indicator=True)

# Select rows from old_data that are not in merged_df
unmatched_rows = merged[merged['_merge'] == 'left_only'].drop(columns=['_merge'])
```

```python
unmatched_rows.to_csv("changed_data.csv", index=False)
```

```python
# Convert CAS values to lists and count the number of CAS values
correct_data['CAS'] = correct_data['CAS'].str.split(', ')
old_data['CAS'] = old_data['CAS'].str.split(', ')
correct_data['CAS_count'] = correct_data['CAS'].apply(len)
old_data['CAS_count'] = old_data['CAS'].apply(len)

# Select rows from 'correct_data' where CAS count is greater than 'old_data'
more_cas_df = correct_data[correct_data['CAS_count'] > old_data['CAS_count']]

# Select rows from 'correct_data' where CAS count is less than 'old_data'
less_cas_df = correct_data[correct_data['CAS_count'] < old_data['CAS_count']]

# Select rows from 'correct_data' where CAS count is the same, but CAS values are different
same_cas_diff_values_df = correct_data[(correct_data['CAS_count'] == old_data['CAS_count']) & 
                                       (correct_data['CAS'] != old_data['CAS'])]

# Drop the CAS_count columns
more_cas_df = more_cas_df.drop(columns=['CAS_count'])
less_cas_df = less_cas_df.drop(columns=['CAS_count'])
same_cas_diff_values_df = same_cas_diff_values_df.drop(columns=['CAS_count'])
```

```python
# Convert CAS values to lists and count the number of CAS values
correct_data['Deprecated CAS'] = correct_data['Deprecated CAS'].str.split(', ')
old_data['Deprecated CAS'] = old_data['Deprecated CAS'].str.split(', ')
correct_data['DEP_count'] = correct_data['Deprecated CAS'].apply(len)
old_data['DEP_count'] = old_data['Deprecated CAS'].apply(len)

# Select rows from 'correct_data' where CAS count is greater than 'old_data'
more_dep_df = correct_data[correct_data['DEP_count'] > old_data['DEP_count']]

# Select rows from 'correct_data' where CAS count is less than 'old_data'
less_dep_df = correct_data[correct_data['DEP_count'] < old_data['DEP_count']]

# Select rows from 'correct_data' where CAS count is the same, but CAS values are different
same_dep_diff_values_df = correct_data[(correct_data['DEP_count'] == old_data['DEP_count']) & 
                                       (correct_data['Deprecated CAS'] != old_data['Deprecated CAS'])]

# Drop the CAS_count columns
more_dep_df = more_dep_df.drop(columns=['DEP_count'])
less_dep_df = less_dep_df.drop(columns=['DEP_count'])
same_dep_diff_values_df = same_dep_diff_values_df.drop(columns=['DEP_count'])
```
