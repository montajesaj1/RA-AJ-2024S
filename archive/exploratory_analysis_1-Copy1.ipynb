```python
import pandas as pd
```

```python
pharm_df = pd.read_csv("scraped_with_both_synonyms.csv")
```

```python
contains_bioassay = pharm_df['Link'].str.contains('bioassay', case=False, na=False)
```

```python
count_bioassay = contains_bioassay.sum()
```

```python
pharm_df_bioassay = pharm_df[contains_bioassay]
```

```python
pharm_df_bioassay.describe()
```

```python
pharm_df_substance = pharm_df[pharm_df["Compound/Substance"] == "SUBSTANCE"]
pharm_df_substance.describe()
```
