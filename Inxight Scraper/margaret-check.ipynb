```python
import pandas as pd
```

```python
inxight = pd.read_csv("inxight_data_final.csv")
inxight.head()
```

```python
margaret = pd.read_excel("inxight_data_Margaret.xlsx")
margaret.head()
```

```python
margaret_cas_url = margaret[['url', 'preferred', 'cas']]
inxight_cas_url = inxight[['cas', 'best_result_url', 'parsed_molecule']]
```

```python
margaret_cas_url
```

```python
merged_cas = pd.merge(inxight_cas_url, margaret_cas_url, how='left', on='cas')
merged_cas
```

```python
matched_cas = merged_cas[merged_cas['best_result_url'] == merged_cas['url']]
matched_cas
```

```python
unmatched_cas = merged_cas[merged_cas['best_result_url'] != merged_cas['url']]
unmatched_cas
```
